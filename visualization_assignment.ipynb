```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Line Plot — Ranking ATP de Ben Shelton
Este gráfico muestra la evolución simulada del ranking ATP de Ben Shelton a lo largo de 2025. Se eligió un line plot porque permite visualizar claramente la tendencia de una variable a lo largo del tiempo.Los datos son simulados en base a tendencias reales del jugador durante ese período.


```python
# Ranking ATP de Ben Shelton - 2025
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

ranking = [18, 16, 15, 14, 17, 15, 13, 12, 11, 13, 12, 10]

plt.figure(figsize=(10, 5))
plt.plot(meses, ranking, color='royalblue', marker='o', linestyle='-', linewidth=2, markersize=7)
plt.title('Evolución del Ranking ATP de Ben Shelton (2025)')
plt.xlabel('Mes')
plt.ylabel('Posición en el ranking ATP')
plt.gca().invert_yaxis()  # Invierte el eje porque menor número = mejor ranking
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 2. Bar Chart — Calorías Quemadas por Actividad
Este gráfico compara las calorías quemadas por hora en distintas actividades físicas. Se eligió un bar chart porque es ideal para comparar valores entre categorías discretas. Los datos son aproximaciones promedio basadas en referencias de salud deportiva.


```python
# Calorías quemadas por actividad física
actividades = ['Correr', 'Ciclismo', 'Natación', 'Fútbol', 'Yoga', 'Pesas']
calorias = [600, 500, 550, 480, 200, 350]

plt.figure(figsize=(10, 5))
plt.bar(actividades, calorias, color=['tomato', 'steelblue', 'mediumseagreen', 'orange', 'mediumpurple', 'goldenrod'])
plt.title('Calorías Quemadas por Hora según Actividad Física')
plt.xlabel('Actividad')
plt.ylabel('Calorías (kcal/hora)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('imgs/bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 3. Box Plot — Variación de Temperatura por Estación
Este gráfico muestra la distribución de temperaturas diarias simuladas para cada estación del año.
Se eligió un box plot porque permite visualizar la dispersión, la mediana y los valores atípicos de cada grupo.
Los datos fueron generados con NumPy simulando condiciones climáticas típicas.

```python
# Temperaturas simuladas por estación
np.random.seed(42)
verano = np.random.normal(30, 4, 90)
otono = np.random.normal(18, 3, 90)
invierno = np.random.normal(8, 5, 90)
primavera = np.random.normal(22, 4, 90)

datos = [verano, otono, invierno, primavera]
estaciones = ['Verano', 'Otoño', 'Invierno', 'Primavera']

plt.figure(figsize=(10, 5))
plt.boxplot(datos, tick_labels=estaciones, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='steelblue'),
            medianprops=dict(color='red', linewidth=2))
plt.title('Variación de Temperatura por Estación del Año')
plt.xlabel('Estación')
plt.ylabel('Temperatura (°C)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('imgs/box_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 4. Histogram — Distribución de Temperaturas Mensuales
Este gráfico muestra la distribución de temperaturas promedio mensuales simuladas a lo largo de varios años.
Se eligió un histograma porque permite visualizar cómo se distribuye una variable continua en rangos.
Los datos fueron generados con NumPy simulando un clima templado.

```python
# Temperaturas mensuales simuladas
np.random.seed(7)
temperaturas = np.random.normal(18, 7, 200)

plt.figure(figsize=(10, 5))
plt.hist(temperaturas, bins=20, color='coral', edgecolor='black', alpha=0.8)
plt.title('Distribución de Temperaturas Promedio Mensuales')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5. Scatter Plot — Aces por Partido vs Ranking ATP
Este gráfico explora la relación entre la cantidad de aces promedio por partido y la posición en el ranking ATP de distintos jugadores.
Se eligió un scatter plot porque es ideal para visualizar la relación entre dos variables continuas.
Los datos son simulados con valores verosímiles basados en estadísticas reales del circuito ATP.

```python
# Aces vs Ranking ATP - jugadores simulados
np.random.seed(15)
jugadores = ['Shelton', 'Isner', 'Karlovic', 'Djokovic', 'Alcaraz',
             'Sinner', 'Medvedev', 'Zverev', 'Ruud', 'Fritz',
             'Hurkacz', 'Rublev', 'Tsitsipas', 'Paul', 'De Minaur']

aces = [8.5, 12.0, 14.5, 3.2, 4.1, 3.8, 5.0, 6.3, 2.9, 7.8,
        9.1, 4.5, 3.5, 6.0, 2.1]
ranking = [12, 45, 80, 2, 3, 1, 5, 4, 8, 10,
           9, 7, 11, 13, 15]

plt.figure(figsize=(10, 6))
plt.scatter(aces, ranking, color='darkorange', edgecolors='black', s=100, alpha=0.8)

for i, nombre in enumerate(jugadores):
    plt.annotate(nombre, (aces[i], ranking[i]), textcoords='offset points', xytext=(5, 5), fontsize=8)

plt.gca().invert_yaxis()
plt.title('Relación entre Aces por Partido y Ranking ATP')
plt.xlabel('Aces promedio por partido')
plt.ylabel('Posición en el Ranking ATP')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('imgs/scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
